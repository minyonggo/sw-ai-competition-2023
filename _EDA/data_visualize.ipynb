### Validation set   vs.  Test set images

```python
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from glob import glob
import math
```

```python
def compare_random_images(image_paths_list1, image_paths_list2, images_per_list=4):
    # Randomly select images_per_list image paths from each list
    random_image_paths1 = random.sample(image_paths_list1, images_per_list)
    random_image_paths2 = random.sample(image_paths_list2, images_per_list)

    # Create the figure with subplots
    fig, axs = plt.subplots(2, 4, figsize=(12, 10))

    # Iterate over the subplots and load/display the images from the first list
    for i in range(2):
        for j in range(2):
            img = mpimg.imread(random_image_paths1[i * 2 + j])
            axs[i, j].imshow(img)
            axs[i, j].set_title(f'Image {i * 2 + j + 1}')

    # Create a second set of subplots for the images from the second list
    for i in range(2):
        for j in range(2):
            img = mpimg.imread(random_image_paths2[i * 2 + j])
            axs[i, j + 2].imshow(img)
            axs[i, j + 2].set_title(f'Image {i * 2 + j + 1 + images_per_list}')

    # Show the comparison plot
    plt.tight_layout()
    plt.show()

# Assuming you have two lists of image paths called 'image_paths_list1' and 'image_paths_list2'

val_list = glob("/home/mykang/mmsegmentation/data/Satellite/img_dir/val_slice/*.png")
test_list = glob("/home/mykang/mmsegmentation/data/Satellite/img_dir/test/*.png")

compare_random_images(val_list, test_list)
```
